```python
# Import
import mysql.connector 
# ---------------------------------------------------------- 
# Connect to DB
# ---------------------------------------------------------- 
space = "__"
# ---------------------------------------------------------- 
host="localhost"
user="root" 
database="pr_mmbuzz_2022_06"
mydb = mysql.connector.connect(host=host,user=user,password="",database=database)
mycursor = mydb.cursor()
# ---------------------------------------------------------- 
query = " Select "
# ---------------------------------------------------------- 
query += " teamapi_id  "  
query += " , name  "  
# ---------------------------------------------------------- 
query += " from football_teams "   
query += " where logo is null "   
query += " order by id asc "  
# ----------------------------------------------------------  
mycursor = mydb.cursor()
mycursor.execute(query)
result =  mycursor.fetchall()
# ----------------------------------------------------------  
total_rows = len(result)
print(space + "Total Row(s) : " + str(total_rows), flush=True) 
# ----------------------------------------------------------  
counter = 0
# ----------------------------------------------------------  
space += "__"
# ---------------------------------------------------------- 
for x in result:   
    # ------------------------------------------------------
    counter += 1
    # ------------------------------------------------------
    teamapi_id = str(x[0])
    name = str(x[1])
    # ------------------------------------------------------total_rows
    word = space + "[" + str(counter) + "/" + str(total_rows) + "] "
    word += " #" + teamapi_id
    word += " " + name
    print(word, flush=True)
    # ------------------------------------------------------
    query1 = " Select "
    # ------------------------------------------------------
    query1 += " teams_home_logo  "   
    # ------------------------------------------------------
    query1 += " from football_fixturesx "   
    query1 += " where teams_home_id = '"+teamapi_id+"' "  
    query1 += " limit 1 "  
    # ------------------------------------------------------
    mycursor = mydb.cursor()
    mycursor.execute(query1)
    result1 =  mycursor.fetchall()
    # ------------------------------------------------------
    total_rows1 = len(result1)
    print(space + "__Total Row(s) 1 : " + str(total_rows1), flush=True) 
    # ------------------------------------------------------ 
    for y in result1:   
        # -------------------------------------------------- 
        teams_home_logo = str(y[0])
        # -------------------------------------------------- 
    # ------------------------------------------------------
    update = " UPDATE `football_teams` SET  "
    update += " `logo`='"+teams_home_logo+"'  "
    update += " where `teamapi_id` = '"+teamapi_id+"'  "
    # ------------------------------------------------------
    print("___---------------------------------------- update", flush=True)
    mycursor.execute(update)
    mydb.commit()  
    # ------------------------------------------------------
    # ------------------------------------------------------
```
