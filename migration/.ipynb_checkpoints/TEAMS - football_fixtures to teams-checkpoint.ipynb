```python
query = ' select '
query += ' teams_home_id ' 
query += ' from football_fixtures ' 
query += ' where teams_home_id is not null ' 
query += ' group by teams_home_id ' 
query += ' order by teams_home_id asc ' 

mycursor = mydb.cursor() 
mycursor.execute(query)
myresult = mycursor.fetchall()

print("Total rows : " + str(len(myresult)))

for x in myresult:  
    teamapi_id = str(x[0])
    print(teamapi_id + " -- " + str(len(teamapi_id)) )
    
    query_1 = ' select '
    query_1 += ' * ' 
    query_1 += ' from teams ' 
    query_1 += ' where teamapi_id = "'+teamapi_id+'" '  
    
    mycursor = mydb.cursor() 
    mycursor.execute(query_1)
    myresult_1 = mycursor.fetchall()
    
    len_myresult_1 = len(myresult_1)
    
    if(len_myresult_1 == 0):
        insert = " INSERT INTO `teams`(`teamapi_id` " 
        insert += " ) VALUES ( "
        insert += " '"+teamapi_id+"'  "
        insert += " )  "
        mycursor.execute(insert)
        mydb.commit()  
        print("Row Inserted")
    
    
```

```python
import mysql.connector
# -----------------------------------------------------------------
# UPDATE NAME name from teams
# -----------------------------------------------------------------
host="localhost"
user="root"
database="pr_mmbuzz_2022_06" 
mydb = mysql.connector.connect( host=host, user=user, password="", database=database )
# -----------------------------------------------------------------
query = ' select '
query += ' teamapi_id ' 
query += ' , name ' 
query += ' from teams '  
query += ' where name is null ' 
query += ' order by teamapi_id asc ' 
# -----------------------------------------------------------------
mycursor = mydb.cursor() 
mycursor.execute(query)
myresult = mycursor.fetchall()
# -----------------------------------------------------------------
print("Total rows : " + str(len(myresult))) 
# -----------------------------------------------------------------
for x in myresult:  
    # -------------------------------------------------------------
    teamapi_id = str(x[0])
    teams_name = str(x[1])
    # -------------------------------------------------------------
    print(teamapi_id + " -- " + teams_name ) 
    # -------------------------------------------------------------
    query_1 = ' (select '
    query_1 += ' teams_home_id as id ' 
    query_1 += ' , teams_home as nama ' 
    query_1 += ' from football_fixtures ' 
    query_1 += ' where teams_home_id = "'+str(teamapi_id)+'" ' 
    query_1 += ' group by teams_home_id, teams_home )  ' 
    query_1 += ' union  ' 
    query_1 += ' (select '
    query_1 += ' teams_away_id as id ' 
    query_1 += ' , teams_away as nama ' 
    query_1 += ' from football_fixtures ' 
    query_1 += ' where teams_away_id = "'+str(teamapi_id)+'" ' 
    query_1 += ' group by teams_away_id, teams_away )  ' 
    # -------------------------------------------------------------
    mycursor = mydb.cursor() 
    mycursor.execute(query_1)
    myresult_1 = mycursor.fetchall()
    # -------------------------------------------------------------
    len_myresult_1 = len(myresult_1)
    # -------------------------------------------------------------
    print("  football_fixtures Total rows : " + str(len_myresult_1)) 
    # -------------------------------------------------------------
    if(len_myresult_1 == 1):
        # ---------------------------------------------------------
        for y in myresult_1:  
            # -----------------------------------------------------
            teamapi_id_y = str(y[0])
            teams_name_y = str(y[1])
            # -----------------------------------------------------
        # ---------------------------------------------------------
        update = " UPDATE `teams` SET "
        update += " `name`= '"+teams_name_y.replace("'", "\\'")+"' "
        update += " where `teamapi_id`= '"+teamapi_id_y+"' "
        # ---------------------------------------------------------
        mycursor.execute(update)
        mydb.commit()  
        print("    Row updated")
        # --------------------------------------------------------- 
    elif(len_myresult_1 != 1):
        # ---------------------------------------------------------
        for y in myresult_1:  
            # -----------------------------------------------------
            teamapi_id_y = str(y[0])
            teams_name_y = str(y[1])
            # -----------------------------------------------------
            print("#" + teamapi_id_y + " -- " + teams_name_y)
        # ---------------------------------------------------------
    # -------------------------------------------------------------
         
    
    
```
